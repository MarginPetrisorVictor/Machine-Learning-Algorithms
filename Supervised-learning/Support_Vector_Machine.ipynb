## **Support Vector Machine (SVM)**

This method is a highly mathematical one with the core idea derived from mathematical vectors. To be able to categorize a dataset, we start by defining a vector $\vec{w}$ that would represent a perpendicular vector to the decision boundary that we will find later. We can make the condition that the vector that points from the origin to the data point multiplied by $\vec{w}$ (we use dot product because we are working with vectors) has to be bigger than a constant $c$ to be considered a positive sample. 
$$ \vec{w}\cdot\vec{x} \ge c $$

Without taking the generalization, we can transform the inequation into the following form:
$$ \vec{w}\cdot\vec{x} + b \ge 0 $$

We also want to define a lower margin and an upper margin that respect the following inequations, in order to separate the samples:
$$ \vec{w}\cdot\vec{x_+} + b \ge 1 $$
$$ \vec{w}\cdot\vec{x_-} + b \le -1 $$

This two margins represent the street between the two categories that are present in the dataset. Our mission is to maximize the width of this street. To make the problem more convinient we use $y_i = 1$ for positive samples and $y_i = -1$ for negative samples. If we use the notation in our inequations defined above we obtain the following result: 
$$ y_i(\vec{w}\cdot\vec{x_i} + b) - 1 \ge 0$$

The gutters of the streets will respect the following equation:
$$ y_i(\vec{w}\cdot\vec{x_i} + b) - 1 = 0$$

Now that we defined the margins for our street we can calculate the width as follows:
$$ width = (\vec{x_+} - \vec{x_-})\cdot\frac{\vec{w}}{||w||} $$

Using the equation that we defined above, we can exchange $x_+$ and $x_-$ so that the new width equation is:
$$ width = \frac{2}{||w||} $$

As stated previously, we want to maximize the width of the street and this statement can be rephrased as minimizing the following function:
$$ min(\frac{1}{2}||w||^2)$$

This is the part where it get's a little complicated. In a normal case we would just calculate the derivative and set that to 0. In this case, because we work with the constraints, we must formulate the Lagrangean, then we can calculate the partial derivatives. The formula for the Lagrangean is:
$$ L(w, b, \alpha) = \frac{1}{2}||w||^2 - \sum_{i=1}^{m}\alpha_i[y_i(\vec{w}\cdot\vec{x_i} + b) - 1]$$

Now we can compute the partial derivatives with regards to w and b:
$$ \frac{dL}{dw} = \vec{w} - \sum_{i=1}^{m}\alpha_iy_i\vec{x_i} = 0$$
$$ \frac{dL}{db} = -\sum_{i=1}^{m}\alpha_iy_i = 0 $$

Here we can solve the optimization problem by exchanging **w** in the Lagrangean and also consider the equation $\sum_{i=1}^{m}\alpha_iy_i = 0 $ as a new constraint. The new form is called the dual form of the Lagrangean and for this form we need to maximize the following equation:
$$ L_d = -\frac{1}{2}\sum_{i=1}^m\sum_{j=1}^{m}\alpha_i\alpha_jy_iy_j\vec{x_i}\vec{x_j} + \sum_{i=1}^{m}\alpha_i $$

## **Quadratic Programming**

In order for us to optimize the result we obtained before we use the quadratic programming. A classical statement of the quadratic programming is the following:
$$ min(\frac{1}{2}x^TQx + c^Tx) $$
$$ Ax = b $$
$$ Dx \le g $$
Using the algorithm named active set method we can solve the most optimal sets of x in order to minimize the quadratic function

```python
import numpy as np
import matplotlib.pyplot as plt
from cvxopt import matrix, solvers

def generate_linear_data(n=100):
    X = np.random.randn(n, 2)
    y = np.sign(X[:, 0] + X[:, 1]) 
    return X, y

X_linear, y_linear = generate_linear_data()
```

```python
def linear_kernel(X):
    return np.dot(X, X.T)

def support_vector_machine_linear(X, y, C = 10, tol = 1e-6):
    n_samples, n_features = X.shape
    kernel = linear_kernel(X)
    
    Q = matrix(np.outer(y, y) * kernel , tc = 'd')
    q = matrix(-np.ones(n_samples), tc = 'd')
    D = matrix(np.vstack((-np.eye(n_samples), np.eye(n_samples))), tc = 'd')
    g = matrix(np.hstack((np.zeros(n_samples), C*np.ones(n_samples))), tc = 'd')
    A = matrix(y.reshape(1, -1), tc = 'd') 
    b = matrix(0.0, tc = 'd')

    solvers.options['show_progress'] = False
    solution = solvers.qp(Q, q, D, g, A, b)
    alpha = np.ravel(solution['x'])

    w = np.dot((alpha*y), X)
    support_vectors = (alpha > tol) & (alpha < C - tol)
    
    if np.any(support_vectors):
        b = np.mean(y[support_vectors] - np.dot(X[support_vectors], w))
    else:
        b = 0
    
    return w, b, alpha
```

```python
def mean_normalization(X):
    m, n = X.shape
    X_scaled = np.zeros((m,n))
    for i in range(n):
        average = np.mean(X[:, i])
        minimum = np.min(X[:, i])
        maximum = np.max(X[:, i])
        X_scaled[:, i] = (X[:, i] - average)/(maximum - minimum)
    
    return X_scaled
```

```python
def plot_results_linear(X, y):
    x1_min, x1_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    x2_min, x2_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1

    x1, x2 = np.meshgrid(np.linspace(x1_min, x1_max, 500),
                         np.linspace(x2_min, x2_max, 500))
   
    w, b, alpha = support_vector_machine_linear(X, y)
    
    z = w[0] * x1 + w[1] * x2 + b

    plt.contour(x1, x2, z, levels = [-1, 0, 1], linestyles = ['--', '-', '--'], colors = 'k')

    plt.scatter(X[y==1, 0], X[y==1, 1], c = 'b', marker = 'o', label = 'Positive')
    plt.scatter(X[y==-1, 0], X[y==-1, 1], c = 'r', marker = 's', label = 'Negative')

    sv = alpha > 1e-5
    plt.scatter(X[sv, 0], X[sv, 1], s = 100, facecolors = 'none', edgecolors = 'g', label = 'Support Vectors')

    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.title('SVM Decision Boundary and Support Vectors')
    plt.show()
```

```python
X_scaled = mean_normalization(X_linear)
plot_results_linear(X_scaled, y_linear)
```

```python
def generate_nonlinear_data(n=100):
    X = np.random.randn(n, 2)
    y = np.sign(X[:, 0] ** 2 + X[:, 1] ** 2 - 0.5)
    return X, y

X_nonlinear, y_nonlinear = generate_nonlinear_data()
```

```python
def rbf_kernel(X, gamma):
    pairwise_dists = np.sum(X ** 2, axis = 1).reshape(-1, 1) + np.sum(X ** 2, axis = 1) - 2 * np.dot(X, X.T)
    K = np.exp(-gamma * pairwise_dists)
    return K

def poly_kernel(X, coef0, degree):
    K = (np.dot(X, X.T) + coef0) ** degree
    return K

def support_vector_machine_nonlinear(X, y, C = 1, tol = 1e-6, kernel ='rbf', gamma = 0.1, coef0 = 0, degree = 3):
    n_samples, n_features = X.shape

    match kernel:
        case "rbf":
            K = rbf_kernel(X, gamma)
        case "poly":
            K = poly_kernel(X, coef0, degree)

    Q = matrix(np.outer(y, y) * K, tc = 'd')
    q = matrix(-np.ones(n_samples), tc = 'd')
    D = matrix(np.vstack((-np.eye(n_samples), np.eye(n_samples))), tc = 'd')
    g = matrix(np.hstack((np.zeros(n_samples), C*np.ones(n_samples))), tc = 'd')
    A = matrix(y.reshape(1, -1), tc = 'd')
    b = matrix(0.0, tc = 'd')

    solvers.options['show_progress'] = False
    solution = solvers.qp(Q, q, D, g, A, b)
    alpha = np.ravel(solution['x'])

    support_vectors = (alpha > tol) & (alpha < C - tol)
    if np.any(support_vectors):
        bias = np.mean(y[support_vectors] - np.dot(K[support_vectors, :], alpha * y))
    else:
        bias = 0

    return alpha, bias

def decision_function(X_train, X_test, alpha, y, bias, gamma = 0.1):
    K_test = np.exp(-np.sum((X_test[:, np.newaxis] - X_train) ** 2, axis=2) / (2 * gamma**2))
    return np.sum(alpha * y * K_test, axis=1) + bias
```

```python
def plot_results_nonlinear(X, y):
    alpha, bias = support_vector_machine_nonlinear(X, y, kernel = "rbf")

    x1, x2 = np.meshgrid(np.linspace(X[:, 0].min() - 0.1, X[:, 0].max() + 0.1, 100),
                         np.linspace(X[:, 1].min() - 0.1, X[:, 1].max() + 0.1, 100))

    X_test = np.c_[x1.ravel(), x2.ravel()]

    Z = decision_function(X, X_test, alpha, y, bias)
    Z = Z.reshape(x1.shape)

    plt.contourf(x1, x2, Z, levels=np.linspace(Z.min(), Z.max(), 50), cmap="coolwarm", alpha=0.5)
    plt.contour(x1, x2, Z, levels=[0], colors="black", linewidths=2) 
    plt.scatter(X[:, 0], X[:, 1], c=y_nonlinear, cmap="coolwarm", edgecolors="k")
    plt.title("Nonlinear SVM with RBF Kernel Decision Boundary")
    plt.show()
```

```python
X_scaled = mean_normalization(X_nonlinear)
plot_results_nonlinear(X_scaled, y_nonlinear)
```
